```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/aayush1036/data/main/Iris.csv')
data
```

# Scatter plots

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(x='SepalWidthCm', y='PetalLengthCm',data=data)
ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_title('SepalWidthCm vs PetalLengthCm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(x='SepalWidthCm', y='PetalLengthCm',data=data,s=100) #changes the size of the dots 
ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_title('SepalWidthCm vs PetalLengthCm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(x='SepalWidthCm', y='PetalLengthCm',data=data, marker='^') #changes the marker of the points
ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_title('SepalWidthCm vs PetalLengthCm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(x='SepalWidthCm', y='PetalLengthCm',data=data,c='Red') #changes the color of the points
ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_title('SepalWidthCm vs PetalLengthCm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(x='SepalWidthCm', y='PetalLengthCm',data=data, alpha=0.5) #changes the transparency of the points (should be bw 0&1)
ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_title('SepalWidthCm vs PetalLengthCm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(x='SepalWidthCm', y='PetalLengthCm',data=data,c='white', edgecolor='black') 
#changes the color of the edges of the dots
ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_title('SepalWidthCm vs PetalLengthCm')
```

# Line plots

```python
#Creating the points
import numpy as np
x = np.arange(0,101,1)
y = 5*x + 10 # line of form y=5x+10
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Y = 5X + 10')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(x,y, ls='--') # plots dotted line
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Y = 5X + 10')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(x,y, c='black') # plots a black line
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Y = 5X + 10')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(x,y,marker='v') #plots a line connected by triangles
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Y = 5X + 10')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(x,y,linewidth=10) #plots a thicker line
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Y = 5X + 10')
```

# Histograms 

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.hist(data['SepalLengthCm'])
ax.set_title('Iris Dataset Sepal Length in cm')
ax.set_xlabel('Sepal Length in Cm')
ax.set_ylabel('Frequency')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.hist(data['SepalLengthCm'], edgecolor='Black') #separates each bin by a black line
ax.set_title('Iris Dataset Sepal Length in cm')
ax.set_xlabel('Sepal Length in Cm')
ax.set_ylabel('Frequency')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.hist(data['SepalLengthCm'], edgecolor='Black') #separates each bin by a black line
ax.set_title('Iris Dataset Sepal Length in cm')
ax.set_xlabel('Sepal Length in Cm')
ax.set_ylabel('Frequency')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.hist(data['SepalLengthCm'], edgecolor='Black',bins=25, color='Orange') # gives orange color bins
ax.set_title('Iris Dataset Sepal Length in cm')
ax.set_xlabel('Sepal Length in Cm')
ax.set_ylabel('Frequency')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.hist(data['SepalLengthCm'], edgecolor='Black',bins=25,cumulative=True) # gives cumulative histogram
ax.set_title('Iris Dataset Sepal Length in cm')
ax.set_xlabel('Sepal Length in Cm')
ax.set_ylabel('Frequency')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.hist(data['SepalLengthCm'], edgecolor='Black',bins=25,log=True) # assumes logarithmic scale
ax.set_title('Iris Dataset Sepal Length in cm')
ax.set_xlabel('Sepal Length in Cm')
ax.set_ylabel('Frequency')
```

# Boxplots

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.boxplot(x=data['SepalLengthCm']) # gives cumulative histogram
ax.set_title('Iris Dataset Sepal Length in cm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.boxplot(x=data['SepalLengthCm'], notch=True) # produces noteched boxplot
ax.set_title('Iris Dataset Sepal Length in cm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.boxplot(x=data['SepalLengthCm'], vert=False) # gives horizontal boxplot
ax.set_title('Iris Dataset Sepal Length in cm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.boxplot(x=data['SepalLengthCm'], whis=1) 
"""
decides the multiplier for outliers (default = 1.5)
It means that everything below Q1-{whis}*IQR and Q3 + {whis}*IQR will be considered as outliers
"""
ax.set_title('Iris Dataset Sepal Length in cm')
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.boxplot(x=data['SepalLengthCm'], whiskerprops={'linewidth':1.5, 'color':'blue', 'ls':'--'}) # edit the whisker properties
ax.set_title('Iris Dataset Sepal Length in cm')
```

# Bar plots

```python
fig, ax = plt.subplots(figsize=(16,6))
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
```

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N)
width = 0.35

fig, ax = plt.subplots(figsize=(16,6))
ax.bar(ind, menMeans, width, color='r')
ax.bar(ind, womenMeans, width,bottom=menMeans, color='b')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['Men', 'Women'])
```

```python
python = (85, 67, 23, 98)
java = (50, 67, 89, 14)
networking = (60, 20, 56, 22)
machine_learning = (88, 23, 40, 87)

people = ["Bob", "Anna", "John", "Mark"]

index = np.arange(4)

ig, ax = plt.subplots(figsize=(16, 6))
ax.bar(index, python, width=0.2, label="Python")  # Python is the first bar so offset is 0
ax.bar(index + 0.2, java, width=0.2, label="Java")  # Java is just after python so offset is 0.2
ax.bar(index + 0.4, networking, width=0.2, label="Networking")  # Networking is just after java so offset = 0.2+0.2
ax.bar(index + 0.6, machine_learning, width=0.2, label="ML")   # ML is just after networking so offset = 0.2+0.2+0.2
ax.set_title("IT Skill Levels")
ax.set_xlabel("Person")
ax.set_ylabel("Level")
ax.legend(loc='upper left')
ax.set_ylim(0, 120) # to accomodate the legend
plt.xticks(index + 0.2, people);
```

# Add axes

```python
x = np.linspace(1,50,50)
y=x**2
fig = plt.figure(figsize=(16,6))
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('x=y^2')
axes1.set_title('y=x^2');
```
