# Cluster map

```
Heirarchically clustered heatmap
Distances between the points are going to be calculated and the closest ones are going to be joined
This is going to be continued for the next closest and it is going to then compare the columns and the rows of our heatmap
```

````python
iris = sns.load_dataset('iris')
species = iris.pop('species')
sns.clustermap(iris)
````

````python
sns.clustermap(flights, cmap='Blues',standard_scale=1)
````

# Pairgrid

````python
iris = sns.load_dataset('iris')
iris_grid = sns.PairGrid(iris, hue='species') #common args for every plot
iris_grid.map(plt.scatter)
````

````python
iris = sns.load_dataset('iris')
iris_grid = sns.PairGrid(iris, hue='species')
iris_grid.map_diag(plt.hist)
iris_grid.map_offdiag(plt.scatter)
````

````python
iris = sns.load_dataset('iris')
iris_grid = sns.PairGrid(iris, hue='species')
iris_grid.map_diag(plt.hist)
iris_grid.map_upper(plt.scatter)
iris_grid.map_lower(sns.kdeplot)
````

````python
#Custom plot
iris = sns.load_dataset('iris')
iris_grid = sns.PairGrid(iris, hue="species", x_vars = ["sepal_length", 'sepal_width'],y_vars = ['petal_length', 'petal_width'])
iris_grid.map(plt.scatter)
iris_grid.add_legend()
````

# Facet Grid

````python
tips_fg = sns.FacetGrid(tips_data, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
````

````python
tips_fg = sns.FacetGrid(tips_data, col='time', row='smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
````

````python
tips_fg = sns.FacetGrid(tips_data, col='time', hue='smoker', height=4, aspect=1.3)
tips_fg.map(plt.scatter, 'total_bill', 'tip')
````

````python
tips_fg = sns.FacetGrid(tips_data, col='time', hue='smoker', height=4, aspect=1.3,
                        col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
````

````python
kws=dict(s=50, linewidth=0.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_data, col='sex', hue='smoker', height=4, aspect=1.3,
                       hue_order=['Yes', 'No'], hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
````

````python
att_df = sns.load_dataset('attention')
att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height=1.5)
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
````
