```python
import numpy as np
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
```

```python
iris = load_iris()
```

# Converting to dataframe (optional)

```python
iris_x = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_x.head()
```

```python
iris_y = pd.DataFrame(data=iris.target, columns=['species'])
```

```python
combined_iris = pd.concat([iris_x,iris_y],axis=1)
```

```python
combined_iris.head()
```
